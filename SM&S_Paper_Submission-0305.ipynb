# Finding the most frequent hashtags from each community

```python
df = pd.read_csv('/Users/ymn/Box/usher/data_manipulated/raw/all_filtered_with_hashtags_mentions_columns.csv') #File is not shared on GitHub to comply with Twitter policy
```

```python
# Example of Community 1
df_retweet_bwtween = df[df['user_screen_name'].isin(df_community_1_people) & df['retweet_or_quote_screen_name'].isin(df_community_1_people)]
df_reply_bwtween = df[df['user_screen_name'].isin(df_community_1_people) & df['in_reply_to_screen_name'].isin(df_community_1_people)]
df_original_between = df[(df['user_screen_name'].isin(df_community_1_people))& (df['tweet_type'] == 'original')]
                        
result = pd.concat([df_retweet_bwtween,df_reply_bwtween,df_original_between])
```

```python
list_0 = []
for i in result['hastages_more'].values:
    if str(i)!='nan':
        combine = str(i).split(',')
        for c in range(len(combine)):
            list_0.append(combine[c].lower().replace("'","").replace("[","").replace("]","").replace(" ",""))
```

```python
result['hastages_more']
```

```python
from collections import Counter
Counter(list_0).most_common(50)
```

# Finding the most frequent mentions from each community

```python
df_orig = pd.read_csv('/Users/ymn/Box/usher/raw/all_filtered.csv') # File is not shared on GitHub to comply with Twitter policy
unique_name = list(df_orig['user_screen_name'].unique())
```

```python
# Example of community 19
df_retweet_bwtween = df[df['user_screen_name'].isin(df_community_19_people) & df['retweet_or_quote_screen_name'].isin(df_community_19_people)]
df_reply_bwtween = df[df['user_screen_name'].isin(df_community_19_people) & df['in_reply_to_screen_name'].isin(df_community_19_people)]
df_original_between = df[(df['user_screen_name'].isin(df_community_19_people))& (df['tweet_type'] == 'original')]
                        

result = pd.concat([df_retweet_bwtween,df_reply_bwtween,df_original_between])
```

```python
result['mentions_more']
```

```python
mentions_ = []
for i in result['mentions_more'].values:
    if str(i)!='nan':
        combine = str(i).split(',')
        for c in range(len(combine)):
            name = combine[c].replace("'","").replace("[","").replace("]","").replace(" ","")
            if name not in unique_name:
                mentions_.append(name)
```

```python
from collections import Counter
Counter(mentions_).most_common(50)
```
